# Code to extract winners in Mexico GP in Autodromo Hermanos Rodriguez

This code is created to make a DataFrame with the winners of the Mexico Grand Prix in Autodromo Hermanos Rodriguez. The data is taken from the page of pitwall.app

## Setup

Let's load some libraries

```python
import pandas as pd
import numpy as np
```

## Code

The race was held in the years from 1963 to 1970, from 1986 to 1992, from 2015 to 2019 and 2021.

```python
years = [2019, 2018, 2017, 2016, 2015, 
         1992, 1991, 1990, 1989, 1988, 
         1987, 1986, 1970, 1969, 1968,
         1967, 1966, 1965, 1964, 1963]
```

The following code will scrap from the pitwall.app to create a DataFrame with the winners of the Singapore Grand Prix

```python
read_data = pd.read_html('https://pitwall.app/races/2021-mexico-city-grand-prix')
results = read_data[0].drop(columns=['Pos','Pts'])
winner = results.loc[[0]].assign(Year=2021)
winner.insert(5, column='Delta', value=float(results.iat[1,2].strip('+')))
winner.insert(5, column='Second', value=results.iat[1,0])
winner_year = winner.set_index('Year')
for y in years:
    print(y)
    url = str(y)
    url = 'https://pitwall.app/races/'+ url + '-mexican-grand-prix'
    try:
        read_data = pd.read_html(url)
    except:
        continue
    results = read_data[0].drop(columns=['Pos','Pts'])
    winner = results.loc[[0]].assign(Year=y)
    winner = winner.set_index('Year')
    try:
        winner.insert(5, column='Delta', value=float(results.iat[1,2].strip('+')))
    except:
        winner.insert(5, column='Delta', value=0)
    winner.insert(5, column='Second', value=results.iat[1,0])
    winner_year = pd.concat([winner_year,winner], axis=0)
winner_year['Driver'] = [x.split(' ', 1)[1].strip() for x in list(winner_year['Driver'])]
winner_year['Second'] = [x.split(' ', 1)[1].strip() for x in list(winner_year['Second'])]
winner_year    
```

```python
winner_year.to_csv('Winner from Mexico Grand Prix.csv')
```
