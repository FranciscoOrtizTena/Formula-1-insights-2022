# Code to extract data from winner in SPA-Francorchamps

## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

## Code

The data is taken from the page pitwall.app

```python
read_data = pd.read_html('https://pitwall.app/races/2021-belgian-grand-prix')
results = read_data[1].drop(columns=['Pos'])
results.loc[[0],['Delta']] = results.iat[1,5]
year = 2021
poleman = results.loc[[0]].assign(Year=year)
poleman_year = poleman.set_index('Year')
i = 1
while i < 20:
    print(year)
    year = year-1
    if year == 2006:
        break
    url = str(year)
    url = 'https://pitwall.app/races/'+ url + '-belgian-grand-prix'
    try:
        read_data = pd.read_html(url)
    except:
        i += 1
        continue
    results = read_data[1].drop(columns=['Pos'])
    results.loc[[0],['Delta']] = results.iat[1,5]
    poleman = results.loc[[0]].assign(Year=year)
    poleman = poleman.set_index('Year')
    poleman_year = pd.concat([poleman_year,poleman], axis=0)
    i += 1
poleman_year    
```

```python
poleman_year.to_csv('poleman year.csv')
```
