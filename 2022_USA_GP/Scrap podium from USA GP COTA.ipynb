# Code to extract the podium data from USA GP in COTA

The following code gives you the list of drivers on the Podium from the USA Grand Prix in the Circuit of the Americas, the info is taken from the web piwall.app

## Setup

First let's load some libraries

```python
import pandas as pd
import numpy as np
```

## Code

The race was held in the years from 2012 to 2019 and in the 2021.

```python
years = [2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012]
```

This will be the 10th race in the circuit of the Americas, so the following code will put all the drivers in the Podium information in one DataFrame

```python
read_data = pd.read_html('https://pitwall.app/races/2021-united-states-grand-prix')
results = read_data[0].drop(columns=['Pts','Time/Retired','Grid','Laps'])
winner = results.loc[[0,1,2]].assign(Year=2021)
winner_pod = winner.set_index('Year')
for y in years:
    print(y)
    url = str(y)
    url = 'https://pitwall.app/races/'+ url + '-united-states-grand-prix'
    try:
        read_data = pd.read_html(url)
    except:
        continue
    results = read_data[0].drop(columns=['Pts','Time/Retired','Grid','Laps'])
    winner = results.loc[[0,1,2]].assign(Year=y)
    winner = winner.set_index('Year')
    winner_pod = pd.concat([winner_pod,winner], axis=0)
winner_pod['Pos'] = [int(x[0]) for x in list(winner_pod['Pos'])]
winner_pod['Driver'] = [x.split(' ', 1)[1].strip() for x in list(winner_pod['Driver'])]
winner_pod
```

Finally let's save it into a CSV file to work on Tableau

```python
winner_pod.to_csv('Drivers Podium USA GP COTA.csv')
```
