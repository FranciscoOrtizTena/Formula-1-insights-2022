# Code to extract data Qualy in Singapore Grand Prix

The following code create a DataFrame with the pole winner of the Qualifying in the Singapore Grand Prix.

## Setup

First lets load some useful libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

## Code

The data is taken from the page pitwall.app

```python
read_data = pd.read_html('https://pitwall.app/races/2019-singapore-grand-prix')
results = read_data[1].drop(columns=['Pos'])
results.loc[[0],['Delta']] = results.iat[1,5]
year = 2019
poleman = results.loc[[0]].assign(Year=year)
poleman_year = poleman.set_index('Year')
i = 1
while i < 20:
    print(year)
    year = year-1
    if year == 2007:
        break
    url = str(year)
    url = 'https://pitwall.app/races/'+ url + '-singapore-grand-prix'
    try:
        read_data = pd.read_html(url)
    except:
        i += 1
        continue
    results = read_data[1].drop(columns=['Pos'])
    results.loc[[0],['Delta']] = results.iat[1,5]
    poleman = results.loc[[0]].assign(Year=year)
    poleman = poleman.set_index('Year')
    poleman_year = pd.concat([poleman_year,poleman], axis=0)
    i += 1
poleman_year    
```

```python
poleman_year.to_csv('poleman Singapore GP Qualy.csv')
```
