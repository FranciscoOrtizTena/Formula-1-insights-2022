# Code to extract winners in Brazilian GP in Autodromo José Carlos Pace

This code is created to make a DataFrame with the winners of the Brazilian Grand Prix in Autodromo José Carlos Pace. The data is taken from the page of pitwall.app

## Setup

Let's load some libraries

```python
import pandas as pd
import numpy as np
```

## Code

The race was held in the years from 1973 to 1977, 1979, 1980 and from 1990 to 2021, without 2020 due to the pandemic

```python
years = [2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012, 
         2011, 2010, 2009, 2008, 2007, 2006, 2005, 2004,
         2003, 2002, 2001, 2000, 1999, 1998, 1997, 1996,
         1995, 1994, 1993, 1992, 1991, 1990, 1980, 1979, 
         1977, 1976, 1975, 1974, 1973]
```

The following code will scrap from the pitwall.app to create a DataFrame with the winners of the Singapore Grand Prix

```python
read_data = pd.read_html('https://pitwall.app/races/2021-brazilian-grand-prix')
results = read_data[0].drop(columns=['Pos','Pts'])
winner = results.loc[[0]].assign(Year=2021)
winner.insert(5, column='Delta', value=float(results.iat[1,2].strip('+')))
winner.insert(5, column='Second', value=results.iat[1,0])
winner_year = winner.set_index('Year')
for y in years:
    print(y)
    url = str(y)
    url = 'https://pitwall.app/races/'+ url + '-brazilian-grand-prix'
    try:
        read_data = pd.read_html(url)
    except:
        continue
    results = read_data[0].drop(columns=['Pos','Pts'])
    winner = results.loc[[0]].assign(Year=y)
    winner = winner.set_index('Year')
    try:
        winner.insert(5, column='Delta', value=float(results.iat[1,2].strip('+')))
    except:
        winner.insert(5, column='Delta', value=0)
    winner.insert(5, column='Second', value=results.iat[1,0])
    winner_year = pd.concat([winner_year,winner], axis=0)
winner_year['Driver'] = [x.split(' ', 1)[1].strip() for x in list(winner_year['Driver'])]
winner_year['Second'] = [x.split(' ', 1)[1].strip() for x in list(winner_year['Second'])]
winner_year    
```

```python
winner_year.to_csv('Winner from Brazilian Grand Prix.csv')
```
