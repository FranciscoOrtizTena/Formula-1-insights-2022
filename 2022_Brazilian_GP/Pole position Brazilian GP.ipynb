# Pole position from Brazilian Grand Prix in Autodromo José Carlos Pace

The following code allows to extract the Pole Position of the Brazilian Grand Prix in Autodromo José Carlos Pace

## Setup

Let's load some libraries

```python
import pandas as pd
import numpy as np
```

## Code

```python
load_data = pd.read_html('https://pitwall.app/circuits/autodromo-jose-carlos-pace')
```

```python
pole_pos = load_data[4].drop(columns=['Winner', 'Date']).drop([0])
pole_pos['Year'] = [x.split(' ', 1)[0] for x in list(pole_pos['Race'])]
pole_pos = pole_pos.drop(columns=['Race']).set_index('Year')
```

```python
pole_pos
```

```python
pole_pos.to_csv('Pole position Mexico GP.csv')
```

```python
pole_pos.to_excel('Pole position Mexico GP.xlsx')
```
