# Code to extract winners in Japanese GP in Suzuka Circuit

This code is created to make a DataFrame with the winners of the Japanese Grand Prix inSuzuka Circuit. The data is taken from the page of pitwall.app

## Setup

Let's load some libraries

```python
import pandas as pd
import numpy as np
```

## Code

The following code will scrap from the pitwall.app to create a DataFrame with the winners of the Japanese Grand Prix

```python
read_data = pd.read_html('https://pitwall.app/races/2019-japanese-grand-prix')
results = read_data[0].drop(columns=['Pos','Pts'])
year = 2019
winner = results.loc[[0]].assign(Year=year)
winner.insert(5, column='Delta', value=float(results.iat[1,2].strip('+')))
winner.insert(5, column='second', value=results.iat[1,0])
winner_year = winner.set_index('Year')
i = 1
while i < 35:
    year = year-1
    print(year)
    if year == 1986:
        break
    url = str(year)
    url = 'https://pitwall.app/races/'+ url + '-japanese-grand-prix'
    try:
        read_data = pd.read_html(url)
    except:
        i += 1
        continue
    results = read_data[0].drop(columns=['Pos','Pts'])
    winner = results.loc[[0]].assign(Year=year)
    winner = winner.set_index('Year')
    try:
        winner.insert(5, column='Delta', value=float(results.iat[1,2].strip('+')))
    except:
        winner.insert(5, column='Delta', value=0)
    winner.insert(5, column='second', value=results.iat[1,0])
    winner_year = pd.concat([winner_year,winner], axis=0)
    i += 1
winner_year['Driver'] = [x.split(' ', 1)[1].strip() for x in list(winner_year['Driver'])]
winner_year['second'] = [x.split(' ', 1)[1].strip() for x in list(winner_year['second'])]
winner_year    
```

```python
winner_year.to_csv('Winner from Japanese Grand Prix.csv')
```
